```r
# Abhishek Krishna 50246436
# Raman Keswani 50246297

# Graph 1 - Influenza National Summary
# File Name: 01_influenza_positive_tests_reported_by_CDC_Clinical_Lab.csv

InfData = read.csv(file = "01_influenza_positive_tests_reported_by_CDC_Clinical_Lab.csv")
a = c()
for(i in 1:length(InfData$Week)){
  a = c(a,InfData$Total.B[i],InfData$Total.A[i])
}
for(i in 19:27){a = c(a,0,0)}
myData = matrix(a, nrow = 2)
names = c(InfData$Week,201806,201808,201810,201812,201814,201816,201818,201820)
namesForXAxis = c(names[1])
for(i in 2:26){
  if(i%%2 == 0){
    namesForXAxis[i] = ""
  }else{
    namesForXAxis[i] = names[i]
  }
}

#length(InfData$Week)
namesForXAxis[27] = names[26]

bp <- barplot(myData, col=c("forestgreen","yellow") , ylim = c(0,19000), axes = FALSE, main = "Influenza Positive Tests Reported to CDC by 
U.S. Clinical Laboratories,\n National Summary 2017-2018 Season")
axis(side = 2, pos = 0, at=seq(0,19000,2000), las =2)
text(bp, -3.7, srt = 60, xpd = TRUE, labels = namesForXAxis , cex=1, adj = c(1,1))
#mtext(text = "Percent Positive", side =1, padj = 1)
legend("topright", legend = c("A","B"), box.lwd = "o", fill = c("yellow","forestgreen"), 
       inset = c(.23,.08), cex = 0.75)
posT = c(InfData$X..Positive,NA,NA,NA,NA,NA,NA,NA,NA,NA)
posB = c(InfData$Percent.Positive.B,NA,NA,NA,NA,NA,NA,NA,NA,NA)
posA = c(InfData$Percent.Positive.A,NA,NA,NA,NA,NA,NA,NA,NA,NA)
par(new = TRUE)
axx=1:26
plot(axx,posT, ylab = "Number of Positive Specimens", col="black", type="l", lwd=2, 
     xlab = "Week", frame.plot = FALSE, axes = FALSE, ylim = c(0.8,30.8))

axis(side = 4,  pos = 26, at=seq(0,30,2), hadj = 0.4, las = 2)
mtext(text = "Percent Positive", side =4, padj = 1.5)
lines(axx,posB, type="l", pch=22, lty=3, lwd=3, col="chartreuse2")
lines(axx,posA, type="l", pch=22, lty=2, lwd=2, col="darkorange")
legend("topright", legend = c("Percent Postive","% Positive Flu A","% Positive Flu B")
       , lwd=2, lty = c(1,2,3), inset = c(.04,.18), box.lwd = "o", cex = 0.75,
       col = c("black", "darkorange","chartreuse2"))
```

```r
# Graph 3: Influenza sub-type pie-charts

#datafile : 03_01_Genetic_04.csv
#datafile2 : 03_02_Genetic_04.csv

datafile = read.csv(file = "03_01_Genetic_04.csv", header = T)
datafile2 = read.csv(file = "03_02_Genetic_04.csv", header = T) 

slices = as.numeric(gsub(',', '', datafile$Data.Cumulative.since.October.1..2017..Week.40.[c(3,2,4,7,6,8)]))
#slices
slices1 = datafile2$Number[c(9,11,10)]
slices2 = datafile2$Number[6]
slices3 = datafile2$Number[1:2]
slices4 = datafile2$Number[3]

lbls1 = slices1
lbls1 = paste(datafile2$Genetic_Group[c(9,11,10)],'\n',lbls1,'\n',datafile2$X..of..Sub.type.Total[c(9,11,10)])

lbls2 = slices2
lbls2 = paste(datafile2$Genetic_Group[6],'\n',lbls2,'\n',datafile2$X..of..Sub.type.Total[6])

lbls3 = slices3
lbls3 = paste(datafile2$Genetic_Group[1:2],'\n',lbls3,'\n',datafile2$X..of..Sub.type.Total[1:2])

lbls4 = slices4
lbls4 = paste(datafile2$Genetic_Group[3],'\n',lbls4,'\n',datafile2$X..of..Sub.type.Total[3])

layout(matrix(c(1,2,2,3,4,5,3,6,7), 3, 3, byrow = TRUE),heights = c(1,2,2), widths = c(1,1,1))

plot.new()
plot.new()
title(sub = "Sequence Results, byGenetic HA Clade/Subclade, of Specimens\nSubmitted to CDC by U.S. Public Health laboratories, Cumulative, \n 2017-2018 Session",
     font.sub = 1)

pie(slices,labels = slices, col=c("red","darkorange","gold","greenyellow","chartreuse3","chartreuse4"), clockwise = TRUE
    ,main="Influenza Positive Specimens Reported by \n U.S. Public Health Laboratories,\n Cumulative, 2017-2018 season", 
        font.main = 1)
legend("bottomleft",legend = sub("^\\s+", "", datafile$Positive.specimens.by.type.subtype[c(3,2,4,7,6,8)]), 
       fill = c("red","darkorange","gold","greenyellow","chartreuse3","chartreuse4"), bty = "n", text.font = 1, text.width =1
      ,cex = 0.75)

pie(slices1,labels = lbls1, col=c("#FB7F7F","#FF1F24","#FCE7E6"), clockwise = TRUE
    ,main="Influenza A(H3N2)", font.main = 1)
box("figure", lty = 1,col = "Black")

pie(slices2,labels = lbls2, col=c("#FFD266"), clockwise = TRUE
    ,main="Influenza A(H1N1)pdm09", font.main = 1)
box("figure",lty = 1,col = "Black")

pie(slices3,labels = lbls3, col=c("darkolivegreen1","white"), clockwise = TRUE
    ,main="Influenza B Victoria", font.main = 1)
box("figure",lty = 1,col = "Black")

pie(slices4, labels = lbls4, col=c("#66C520"), clockwise = TRUE
    ,main="Influenza B Yamagata", font.main = 1)
box("figure",lty = 1,col = "Black")
```

```r
# Graph 4: Mortality

#File Name: 04_NCHSData04.csv

MorData = read.csv(file = "04_NCHSData04.csv")
plot(x = MorData$Percent.of.Deaths.Due.to.Pneumonia.and.Influenza[(c(209:length(MorData$Week)))],
     ylab = "% of All Deaths Due to P&I", col="red", type = "l", cex.main = 0.9,
     xlab = "Week", frame.plot = FALSE, ylim = c(4,12), lwd = "3", xaxt = "n", main = 
    "Pneuomonia and Influenza Mortality from
     the National Center for Health Statistics Mortality Surveillance System")

lines(MorData$Threshold[(c(209:length(MorData$Week)))], type = "l",col="black", lwd ="2")
lines(MorData$Expected[(c(209:length(MorData$Week)))], type = "l",col="black", lwd ="2")

labelsForXAxis = c(40,50,c(10,20,30,40,50),c(10,20,30,40,50),c(10,20,30,40,50),c(10,20,30,40,50))
posForXLabels = c()
for(i in 209:432){
  if(MorData$Week[i] == "40" || MorData$Week[i] == "50" || MorData$Week[i] == "10" || 
     MorData$Week[i] == "20" || MorData$Week[i] == "30"){
        posForXLabels = c(posForXLabels,i-208)
  }
}

par(cex.axis=0.9)
axis(side = 1, labels = labelsForXAxis, at = posForXLabels)
mtext(text = c(2013:2018), side =1, padj = -2, at = c(6, 28, 81, 133, 183,225))
arrows(59,7.4,35,9, lwd = 2, length=.1, code =1)
text(42,9.2,"Epidemic Threshold", cex = 0.75)
arrows(68,7.5,87,5.5, lwd = 2, length=.1, code =1)
text(87,5.35, "Seasonal Baseline", cex = 0.75)
```
